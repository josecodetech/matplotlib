```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Datos de ejemplo
dias = [1,2,3,4,5,6,7]
gasto = [10,25,15,30,35,5,20]
gasto2 = [12,30,43,51,3,32,45]
gasto3 = [21,42,45,64,1,8,24]
```

```python
#Grafica basica
# colores c b g y k w r m
# marker . X o - -. : --
plt.plot(dias,gasto,color='g',marker='o',markersize=4,linewidth=10)
plt.show()
```

```python
plt.plot(dias,gasto,'k-.',markersize=10,linewidth=5)
plt.show()
```

```python
#Grabar grafica
plt.savefig("grafica.png")
```

#Ejes


```python
#limite eje x
plt.xlim(1,4)
#eje y
plt.ylim(1,25)
#cambio de escala
#plt.xscale('log')
#plt.yscale('log')
#a√±adir segundo eje
#plt.twinx()
#no mostrar ejes
#plt.axis('off')
#Invertir ejes
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.plot(dias,gasto)
plt.show()
```

```python
#Etiquetas
plt.title("Gastos")
plt.xlabel("Dias")
plt.ylabel("Gasto")
#plt.xticks(range(1,5,2))
plt.plot(dias,gasto)
plt.legend(('gasto'),loc='upper left')
plt.show()
```

```python
#Multiples graficas en una figura
plt.plot(dias,gasto,color='g')
plt.plot(dias,gasto2,color='r')
plt.plot(dias,gasto3,color='b')
plt.show()
```
