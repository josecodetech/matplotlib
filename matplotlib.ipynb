```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Datos de ejemplo
dias = [1,2,3,4,5,6,7]
gasto = [10,25,15,30,35,5,20]
```

```python
#Grafica basica
# colores c b g y k w r m
# marker . X o - -. : --
plt.plot(dias,gasto,color='g',marker='o',markersize=4,linewidth=10)
plt.show()
```

```python
plt.plot(dias,gasto,'k-.',markersize=10,linewidth=5)
plt.show()
```

```python
#Grabar grafica
plt.savefig("grafica.png")
```
