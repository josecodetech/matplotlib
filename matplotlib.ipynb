```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Datos de ejemplo
dias = [1,2,3,4,5,6,7]
gasto = [10,25,15,30,35,5,20]
gasto2 = [12,30,43,51,3,32,45]
gasto3 = [21,42,45,64,1,8,24]
```

```python
#Grafica basica
# colores c b g y k w r m
# marker . X o - -. : --
plt.plot(dias,gasto,color='g',marker='o',markersize=4,linewidth=10)
plt.show()
```

```python
plt.plot(dias,gasto,'k-.',markersize=10,linewidth=5)
plt.show()
```

```python
#Grabar grafica
plt.savefig("grafica.png")
```

#Ejes


```python
#limite eje x
plt.xlim(1,4)
#eje y
plt.ylim(1,25)
#cambio de escala
#plt.xscale('log')
#plt.yscale('log')
#añadir segundo eje
#plt.twinx()
#no mostrar ejes
#plt.axis('off')
#Invertir ejes
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.plot(dias,gasto)
plt.show()
```

```python
#Etiquetas
plt.title("Gastos")
plt.xlabel("Dias")
plt.ylabel("Gasto")
#plt.xticks(range(1,5,2))
plt.plot(dias,gasto)
plt.legend(('gasto'),loc='upper left')
plt.show()
```

```python
#Multiples graficas en una figura
plt.plot(dias,gasto,color='g')
plt.plot(dias,gasto2,color='r')
plt.plot(dias,gasto3,color='b')
plt.show()
```

#Separando imagen y ejes

```python
fig, ax = plt.subplots()
ax.plot(dias,gasto)
ax.plot(dias,gasto3)
ax.set(xlabel='Dias',ylabel='Gasto',title='Gastos')
ax.legend(['gasto1','gasto3'])
plt.show()
```

```python
#Grafica de ejemplo
fig = plt.figure()
ax=plt.axes()
x = np.linspace(0,20,2000)
ax.plot(x,np.sin(x),color='red',linestyle='dashed')
#alternativa
plt.plot(x,np.sin(x+3),color='blue',linestyle='dotted')
#ajuestes de limites
#plt.xlim(-1.5,10)
#plt.ylim(-2.5,2.5)
#plt.axis('equal')
plt.show()
```

#Subplots, graficas por separado

```python
fig, ax = plt.subplots(3,2,sharey=True,figsize=[9,12])
ax[0][0].plot(dias,gasto,marker='o',color='r',linestyle='--',label='Gastos1')
ax[0][1].plot(dias,gasto2,marker='s',color='b',linestyle='-',label='Gastos2')
ax[1][0].plot(dias,gasto3,marker='D',color='m',linestyle='dotted',label='Gastos3')
ax[0][1].set_title('Gastos 2')
ax[1][0].legend()
ax[0][1].set_xlabel('Dias')
plt.show()
```

#Grafica de barras

```python
datos = pd.DataFrame({'Tienda01': [836,743,936,502],
                      'Tienda02':[776,224,348,522],
                      'Tienda03':[625,775,428,556]},index=('Lunes','Martes','Miercoles','Jueves'))
datos
```

```python
total = datos.sum(axis=1)
total
```

```python
plt.style.use('ggplot')
plt.title("Ventas diarias")
plt.ylabel("Ventas")
plt.xticks(rotation=90)
plt.barh(total.index,total) #o plt.bar
plt.show()
```

#Scatter

```python
x=np.random.randint(10,50,45)
y=np.random.randint(20,81,45)
#Tamaño grafica
plt.figure(figsize=(10,4))
#Crear y posicionar texto
plt.text(9,15,"Esto es un texto nuevo")
plt.scatter(x,y,marker='o',c='red',s=80,alpha=0.90)
plt.title('Numeros al azar')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

```

#Graficas de tarta o pie

```python
ventas = [15000,18000,17000,5000]
tiendas = ['Tienda01','Tienda02','Tienda03','Tienda04']
colores = ['red','gold','blue','green']
plt.pie(ventas,labels=tiendas,colors=colores,labeldistance=1.2,explode=(0.1,0.1,0.1,0.1))
plt.legend(tiendas,loc='lower left')
plt.show()
```
