```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme()
```

```python
#Carga dataset propinas
propinas = sns.load_dataset("tips")
propinas.columns
```

```python
#Visualizamos
sns.relplot(
    data=propinas,
    x='total_bill',y='tip',col='time',
    hue='smoker',style='smoker',size='size'
)
plt.show()
```
