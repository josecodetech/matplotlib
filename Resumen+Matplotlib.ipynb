```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Grafica de linea

```python
x = np.linspace(-np.pi, np.pi, 100)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label='Seno', linestyle='-', color='blue')
plt.plot(x, y2, label='Coseno', linestyle='--', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Funcion seno coseno')
plt.legend()
plt.grid(True)
plt.show()
```

Grafica de dispersion

```python
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
plt.scatter(x, y, color='green', alpha = 0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Grafico de dispersion')
plt.grid(True)
plt.show()
```

Grafica de barras

```python
productos = ['Producto A', 'Producto B']
ventas_enero = [50, 60]
ventas_febrero = [45, 55]
ventas_marzo = [55, 50]
```

```python
x = range(len(productos))
```

```python
plt.bar(x, ventas_enero, width=0.2, label='Enero')
plt.bar([i+0.2 for i in x], ventas_febrero, width=0.2, label='Febrero')
plt.bar([i+0.4 for i in x], ventas_marzo, width=0.2, label='Marzo')
plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.title('Ventas mensuales por productos')
plt.xticks([i+0.2 for i in x], productos)
plt.legend()
plt.grid(True)
plt.show()
```

Histograma

```python
edades = np.random.randint(20, 70, size=100)
```

```python
plt.hist(edades, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma de edades')
plt.grid(True)
plt.show()
```

Grafica de tarta

```python
genero = ['Hombre', 'Mujeres']
porcentaje = [60, 40]
```

```python
plt.pie(porcentaje, labels=genero, autopct='%1.1f%%', colors=['skyblue','pink'], startangle=90)
plt.title('Distribucion de genero')
plt.axis('equal')
plt.show()
```

Grafica de area

```python
dias = np.arange(1,8)
```

```python
temp_max = [25,28,30,32,29,27,26]
temp_min = [15,17,18,20,18,16,15]
```

```python
plt.fill_between(dias, temp_max, temp_min,color='blue',alpha=0.3)
plt.xlabel('Dia')
plt.ylabel('Temperatura')
plt.title('Evolucion de la temperatura semanal')
plt.grid(True)
plt.show()
```

Grafica de violin

```python
np.random.seed(10)
puntos = np.random.normal(70,10,100)
```

```python
plt.violinplot(puntos, vert=False)
plt.xlabel('Puntos')
plt.title('Distribucion de puntos en examen')
plt.grid(True)
plt.show()
```

Grafico de densidad

```python
np.random.seed(0)
datos = np.random.normal(loc=0, scale=1, size=1000)
```

```python
plt.figure(figsize=(8,4))
plt.hist(datos, bins=30, density=True, alpha=0.5, color='blue')
plt.title('Grafico de densidad, KDE')
plt.xlabel('Valor')
plt.ylabel('Densidad')
from scipy.stats import gaussian_kde
kde = gaussian_kde(datos)
x_vals = np.linspace(min(datos),max(datos),1000)
plt.plot(x_vals, kde(x_vals),color='red',linestyle='--',label='KDE')
plt.legend()
plt.grid()
plt.show()
```
